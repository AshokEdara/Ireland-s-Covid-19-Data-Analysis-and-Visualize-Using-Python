Now, let’s group the data based on CountyName. We will plot the data based on the descending order of the number of confirmed cases.

First, we’ll plot a bar chart using the barplot method of the seaborn module.

```python
#imports for plotting.
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
grouped_data = data.groupby('CountyName')['ConfirmedCovidCases', 'ConfirmedCovidDeaths', 'ConfirmedCovidRecovered'].sum().reset_index()
temp = grouped_data.sort_values(by='ConfirmedCovidCases', ascending=False)
data = temp[['CountyName','ConfirmedCovidCases','ConfirmedCovidDeaths','ConfirmedCovidRecovered']]
sns.set_color_codes("muted")
plt.subplots(figsize=(13,7))
sns.barplot(x="ConfirmedCovidCases", y="CountyName", data=data,label="Total", color="green")
```

We got a nice bar chart showing the number of confirmed cases in Ireland. Now, let’s create a pie chart of the same using the pie() method of the matplotlib module.

```python
x=temp['ConfirmedCovidCases']
labels=temp['CountyName']
plt.figure(figsize=(16,10))
plt.title("Ireland's Covid-19 Data Analaysis Using Python", fontsize=6)
plt.pie(x,labels=labels)
```
